```python
#importing libraries 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Dataset
from sklearn import datasets
iris=datasets.load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df.describe()
```

```python
#Checking null values
df.isnull().sum()
```

```python
y=iris.target
print(y)
```

```python
#Training dataset
from sklearn.model_selection import train_test_split
df_train,df_test,y_train,y_test = train_test_split(df,y,test_size = 0.2,random_state = 42)
```

```python
print(df_train.shape)
print(df_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Importing & defining the module


```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df_train,y_train)
```

```python
y_pred = dtree.predict(df_test)
y_pred
```

# Representing it Visually 

```python
from sklearn import tree
fn = ['sepal length(cm)','sepal width(cm)','petal lenght(cm)','petal width(cm)']
cn = ['setosa','versicolour','verginica']

fig, axes = plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi =350)

tree.plot_tree(dtree, feature_names =fn, class_names=cn, filled=True);
```
